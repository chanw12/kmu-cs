```python
import sys
K = df[['people_vaccinated_per_hundred','total_deaths_per_million', 'total_cases_per_million']]
df_new = K.dropna(axis=0)
X = df_new['people_vaccinated_per_hundred']
Y = df_new['total_deaths_per_million']/df_new['total_cases_per_million']


meanx = X.sum()/len(X)
meany = Y.sum()/len(Y)
std_x = (sum((x - meanx) ** 2 for x in X)/len(X))**0.5
std_y = (sum((y-meany)** 2 for y in Y)/len(Y))**0.5

cov = sum((x -meanx) * (y-meany) for x,y in zip(X,Y))/len(X)

corr = cov/std_x.sum()/std_y.sum()
corr
```

```python
import torch
import torch.optim as optim
y_train = torch.FloatTensor(Y).reshape(-1,1)
x_train = torch.FloatTensor(X).reshape(-1,1)

W = torch.zeros(1, requires_grad=True)
b = torch.zeros(1,requires_grad=True)
```

```python
print(x_train.shape)
```

```python
print(y_train.shape)
```

```python
y_train = torch.FloatTensor(Y).reshape(-1,1)
x_train = torch.FloatTensor(X).reshape(-1,1)
W = torch.zeros(1, requires_grad=True)
b = torch.zeros(1, requires_grad=True)

optimizer = optim.SGD([W, b], lr=0.01)

nb_epochs = 20000 
for epoch in range(nb_epochs + 1):

    hypothesis = x_train * W + b
    cost = torch.mean((hypothesis - y_train) ** 2)
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()
    if epoch % 100 == 0:
        print('Epoch {:4d}/{} W: {:.3f}, b: {:.3f} Cost: {:.6f}'.format(
            epoch, nb_epochs, W.item(), b.item(), cost.item()
        ))
```
